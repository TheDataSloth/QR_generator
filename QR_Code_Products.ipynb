```python
import pyqrcode
import pandas as pd
import png
import cv2


products = pd.read_csv('product_list.csv')
products
```

```python
def generateQRCode():
    for index, values in products.iterrows():
        name = values['Name']
        summ = values['Summary']
        quant = values['Quantity']
        itemNum = values['itemNum']
        
        data = f'''
        Name: {name} \n
        Summary: {summ} \n
        Quantity: {quant} \n
        itemNum: {itemNum} \n
        '''
        qr = pyqrcode.create(data)
        qr.png(f"{itemNum}.png", scale='3')
```

```python
generateQRCode()
```
